## **SOAL 1 Modul 1.2**
### RANGE Soal 1

````python
# Dalam suatu liga sepak bola, dipilih secara acak 10 pemain dari klub yang berbeda
# dan diketahui jumlah gol yang mereka raih dalam musim ini adalah: 19, 10, 10, 12, 13, 15, 14, 13, 9, 17.
# Lakukan analisis terhadap sebaran data dari dataset yang diberikan.
# Manakah ukuran sebaran yang paling mampu menggambarkan sebaran data dengan baik?

# Range
# Fungsi untuk menghitung rentang
def calculate_range(data):
  nilai_tertinggi = max(data)
  print(f"Nilai tertinggi: {nilai_tertinggi}")

  nilai_terendah = min(data)
  print(f"Nilai terendah: {nilai_terendah}")

  # Selisih antara nilai tertinggi dan terendah
  data_range = max(data) - min(data)

  print(f"\n")
  print(f"Range: {nilai_tertinggi} - {nilai_terendah} = {data_range}")

  return data_range

# Inisialisasi dataset
dataGol = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
print(f"Data: {dataGol}")

data_range = calculate_range(dataGol)
````

Visualisasi range:

````python
import matplotlib.pyplot as plt

# Create a histogram
plt.hist(dataGol, bins=30, edgecolor='black', alpha=0.5, color='green')

# Define the horizontal line coordinates
y = 1.5 # The y-coordinate where the line will be drawn
x1 = 4/70*1  # The x-coordinate of the start point
x2 = 66/70*1  # The x-coordinate of the end point

# Add a horizontal line
plt.axhline(y=y, xmin=x1, xmax=x2, color='red', linestyle='dashed', linewidth=1, label=f'Range = {calculate_range(dataGol)}')

# Add vertical line
plt.axvline(x=min(dataGol), color='blue', linestyle='dashed', linewidth=1, label=f'Min Data: {min(dataGol):.2f}')
plt.axvline(x=max(dataGol), color='blue', linestyle='dashed', linewidth=1, label=f'Max Data: {max(dataGol):.2f}')

# Add labels and title
plt.xlabel('Jumlah Gol')
plt.ylabel('Frequency')
plt.title('Range')

# Add legend
plt.legend()

# Show the plot
plt.show()
````

### VARIANCE Soal 1

````python
# Fungsi untuk menghitung rata-rata
def calculate_mean(data):
  return sum(data) / len(data)

# Fungsi untuk menghitung variansi
def calculate_variance(data):
  # Rata-rata
  mean = calculate_mean(data)
  print(f"Rata-rata: {mean}")

  # Kuadrat selisih antara setiap nilai data dengan rata-rata
  print(f"Kuadrat selisih setiap nilai data dengan rata-rata:")
  squared_diff = []
  for x in data:
    squared_diff_per_x = (x - mean)**2
    squared_diff.append( squared_diff_per_x )
    print(f"({x} - {mean})**2 = {(x - mean)}**2 = {squared_diff_per_x}")

  print(f"Total kuadrat selisih: {sum(squared_diff)}")
  print(f"Jumlah total data: {len(data)}")

  # Jumlah kuadrat selisih antara setiap nilai data dengan rata-rata, kemudian dibagi dengan jumlah total data
  variance = sum(squared_diff) / len(data)

  print(f"\n")
  print(f"Variansi: {sum(squared_diff)} / {len(data)} = {variance}")

  return variance

print(f"Data: {dataGol}")
variance = calculate_variance(dataGol)
````

### STANDARD DEVIATION Soal 1

````python
# Fungsi untuk menghitung standar deviasi
def calculate_standard_deviation(data):
  variance = calculate_variance(data)
  standard_deviation = variance**0.5
  print(f"\n")
  print(f"Standard Deviation: {variance}**0.5 = {standard_deviation}")

  return standard_deviation

# Inisialisasi dataset
print(f"Data: {dataGol}")

standard_deviation = calculate_standard_deviation(dataGol)
mean = calculate_mean(dataGol)
````

Visualisasi standar deviasi:

````python
import matplotlib.pyplot as plt
import numpy as np

# Buat histogram
plt.hist(dataGol, bins=30, edgecolor='black', alpha=0.5, color='green')

# Tambahkan garis vertikal pada standar deviasi
plt.axvline(x=mean, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean}')
plt.axvline(x=mean - standard_deviation, color='blue', linestyle='dashed', linewidth=1, label=f'{mean-standard_deviation:.2f}')
plt.axvline(x=mean + standard_deviation, color='blue', linestyle='dashed', linewidth=1, label=f'{mean+standard_deviation:.2f}')

# Tambahkan label dan judul
plt.xlabel('Jumlah Gol')
plt.ylabel('Frequency')
plt.title(f'Std Deviation = {standard_deviation:.2f}')

# Tambahkan legenda
plt.legend()

# Tampilkan plot
plt.show()
````

```
Berdasarkan hasil yang didapat, perhitungan menggunakan standar devias
merupakan metode yang paling tepat sebab metode tersebut memperhitungka
seluruh distribusi nilai. Dan juga tak ada outlier yang tidak begitu jauh
membuat standar deviasi kecil
```

### VARIANCE Soal 2

````python
# Fungsi untuk menghitung rata-rata
def calculate_mean(data):
  return sum(data) / len(data)

# Fungsi untuk menghitung variansi
def calculate_variance(data):
  # Rata-rata
  mean = calculate_mean(data)
  print(f"Rata-rata: {mean}")

  # Kuadrat selisih antara setiap nilai data dengan rata-rata
  print(f"Kuadrat selisih setiap nilai data dengan rata-rata:")
  squared_diff = []
  for x in data:
    squared_diff_per_x = (x - mean)**2
    squared_diff.append( squared_diff_per_x )
    print(f"({x} - {mean})**2 = {(x - mean)}**2 = {squared_diff_per_x}")

  print(f"Total kuadrat selisih: {sum(squared_diff)}")
  print(f"Jumlah total data: {len(data)}")

  # Jumlah kuadrat selisih antara setiap nilai data dengan rata-rata, kemudian dibagi dengan jumlah total data
  variance = sum(squared_diff) / len(data)

  print(f"\n")
  print(f"Variansi: {sum(squared_diff)} / {len(data)} = {variance}")

  return variance


# Inisialisasi dataset
print(f"Data: {dataTinggi}")

variance_tinggi = calculate_variance(dataTinggi)
````

### STANDARD DEVIATION Soal 2

````python
# Fungsi untuk menghitung standar deviasi
def calculate_standard_deviation(data):
  variance_tinggi = calculate_variance(data)
  standard_deviation = variance_tinggi**0.5
  print(f"\n")
  print(f"Standard Deviation: {variance}**0.5 = {standard_deviation}")

  return standard_deviation

# Inisialisasi dataset
print(f"Data: {dataTinggi}")

standard_deviation_tinggi = calculate_standard_deviation(dataTinggi)
mean_tinggi = calculate_mean(dataTinggi)
````

Visualisasi standard deviation:

````python
import matplotlib.pyplot as plt
import numpy as np

# Buat histogram
plt.hist(dataTinggi, bins=30, edgecolor='black', alpha=0.5, color='green')

# Tambahkan garis vertikal pada standar deviasi
label1 = mean_tinggi - standard_deviation_tinggi
label2 = mean_tinggi + standard_deviation_tinggi
plt.axvline(x=mean_tinggi, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean_tinggi}')
plt.axvline(x=mean_tinggi - standard_deviation_tinggi, color='blue', linestyle='dashed', linewidth=1, label=f'{label1:.2f}')
plt.axvline(x=mean_tinggi + standard_deviation_tinggi, color='blue', linestyle='dashed', linewidth=1, label=f'{label2:.2f}')

# Tambahkan label dan judul
plt.xlabel('Tinggi Pemain')
plt.ylabel('Frequency')
plt.title(f'Std Deviation = {standard_deviation_tinggi:.2f}')

# Tambahkan legenda
plt.legend()

# Tampilkan plot
plt.show()
````
